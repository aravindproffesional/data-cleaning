```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/Users/aravindkumar/Downloads/20260202 Dataset 2 (v8).xlsx - dataset V8 (1).csv")
```

```python
df= df[df['current_type'] != 'PCE']
df
```

```python
is_claim = df['current_type'] == 'CLAIM'
has_lawfirm = df['claimant_lf'].notna() | df['defendant_lf'].notna()
df = df[~is_claim | has_lawfirm]
```

```python
df
```

```python
# --- Config ---
selected_state  = 'Pennsylvania'   # change state name here
selected_county = None             # set to e.g. 'Allegheny County' or None for all counties
```

```python

# Resolve expense: use expense_paid if present, else fall back to expense_paid_extracted
df['expense'] = df['expense_paid'].combine_first(df['expense_paid_extracted'])
```

```python
# Filter by state and optional county
state_df = df[
    (df['state'] == selected_state) &
    df['claimant_lf'].notna() &
    df['expense'].notna()
]
```

```python
if selected_county:
    state_df = state_df[state_df['county'] == selected_county]
```

```python
# Aggregate: top 15 law firms by total expense
agg = (
    state_df.groupby('claimant_lf')['expense']
    .sum()
    .sort_values(ascending=False)
    .head(15)
)
```

```python
# Plot
location_label = f'{selected_county}, {selected_state}' if selected_county else selected_state
fig, ax = plt.subplots(figsize=(14, 6))
agg.plot(kind='bar', ax=ax, color='darkorange', edgecolor='white')

ax.set_title(f'Top 15 Claimant Law Firms by Expense Paid — {location_label}', fontsize=14)
ax.set_xlabel('Claimant Law Firm', fontsize=11)
ax.set_ylabel('Total Expense Paid ($)', fontsize=11)
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()

print(f'\nState: {selected_state}  |  County: {selected_county or "All"}  |  Law firms shown: {len(agg)}')
```
