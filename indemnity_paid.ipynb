```python
is_claim = df['current_type'] == 'CLAIM'
has_lawfirm = df['claimant_lf'].notna() | df['defendant_lf'].notna()
df = df[~is_claim | has_lawfirm]
```

```python
import matplotlib.pyplot as plt

# --- Config ---
selected_state = 'Pennsylvania'   # change state name here
selected_county = None 

# Resolve indemnity: use indemnity_paid if present, else fall back to indemnity_paid_extracted
df['indemnity'] = df['indemnity_paid'].combine_first(df['indemnity_paid_extracted'])
```

```python
# Filter by state and require claimant_lf + indemnity
state_df = df[
    (df['state'] == selected_state) &
    df['claimant_lf'].notna() &
    df['indemnity'].notna()
]

# if selected_county:
    # state_df = state_df[state_df['county'] == selected_county
```

```python
# Aggregate: top 15 law firms by total indemnity
agg = (
    state_df.groupby('claimant_lf')['indemnity']
    .sum()
    .sort_values(ascending=False)
 
    .head(15)
)
```

```python
# Plot
fig, ax = plt.subplots(figsize=(14, 6))
agg.plot(kind='bar', ax=ax, color='steelblue', edgecolor='white')

ax.set_title(f'Top 15 Claimant Law Firms by Indemnity Paid — {selected_state}', fontsize=14)
ax.set_xlabel('Claimant Law Firm', fontsize=11)
ax.set_ylabel('Total Indemnity Paid ($)', fontsize=11)
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```
